## Analisis

**Sueldo**: El promedio(94k) y la mediana(99k) estan muy alejados. Tambien se obtuvo el desvio estandar, siendo este de 2400, numero muy alto. Como conclusion se observa que esta no es una distribucion normal, ya que en las mismas la median y la media (promedio) son iguales, condicion que no se cumple en este caso.

**Altura**: La media es 1.72 y la mediana 1.75, con un desvio de .27. No es una distribucion normal perfecta, pero si podemos decir que se acerca y se comporta como una.
